````python
import time
Year = 1970
Time = (int(time.time()))
day = Time // 86400
second = Time % 86400
while day >= 365:
    if ( Year % 400 == 0 or ( Year % 4 == 0 and Year % 100 != 0 ) ):
        Year += 1
        day -= 366
    else:
        Year += 1
        day -= 365

print( Year , "年" )
print( second // 3600 + 8, "時" , second % 3600 // 60 ,"分",
      second % 60 % 60 ,"秒",)
````
